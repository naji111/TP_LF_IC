```python
# Importation des bibiotheques importantes
import networkx as nx
import matplotlib.pyplot as plt
```

## 1. Création du réseau sémantique

* On utilise NetworkX pour créer un graphe dirigé (DiGraph)
* On ajoute d'abord tous les concepts comme nœuds
* On définit ensuite les relations entre les concepts sous forme de tuples (source, destination, type_relation)
* Chaque relation est ajoutée comme une arête avec un attribut 'relation'


```python
# 1) Création du réseau sémantique
# Initialisation du graphe dirigé
G = nx.DiGraph()

# Ajout des nœuds (concepts)
concepts = ['chien', 'chat', 'animal', 'viande', 'nourriture', 'être vivant']
G.add_nodes_from(concepts)

# Ajout des relations (arêtes avec labels)
relations = [
    ('chien', 'animal', 'est_un'),
    ('chat', 'animal', 'est_un'),
    ('animal', 'être vivant', 'est_un'),
    ('chien', 'viande', 'mange'),
    ('viande', 'nourriture', 'est_une'),
    ('animal', 'respirer', 'a_besoin_de')
]

# Ajout des arêtes avec leurs attributs
for source, target, relation in relations:
    G.add_edge(source, target, relation=relation)
```

## 2. Visualisation du réseau

* On utilise matplotlib pour créer une figure de taille adaptée
* Le spring_layout permet de positionner les nœuds de façon optimale
* On dessine :
    * Les nœuds en bleu clair
    * Les arêtes en gris avec des flèches 
    * Les labels des nœuds en gras
    * Les labels des relations sur les arêtes
* On désactive les axes et ajoute un titre

```python


# 2) Visualisation du réseau sémantique
plt.figure(figsize=(12, 8))

# Définition du layout pour le placement des nœuds
pos = nx.spring_layout(G, k=1, iterations=50)

# Dessiner les nœuds
nx.draw_networkx_nodes(G, pos, node_color='lightblue', 
                      node_size=2000, alpha=0.7)

# Dessiner les arêtes
nx.draw_networkx_edges(G, pos, edge_color='gray', 
                      arrows=True, arrowsize=20)

# Ajouter les labels des nœuds
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# Ajouter les labels des relations sur les arêtes
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, 
                           font_size=8)

plt.title("Réseau Sémantique", pad=20)
plt.axis('off')
plt.show()
```

## 3. Recherche des connexions

### Pour trouver les relations liées au "chien" :
* On parcourt tous les successeurs du nœud "chien"
* Pour chaque successeur, on récupère le type de relation

### Pour la relation spécifique entre "chien" et "animal" :
* On vérifie l'existence de l'arête
* On récupère et affiche le type de relation

```python

# 3) Recherche des connexions
print("\nRecherche des connexions:")

# Relations associées au concept "chien"
print("\nRelations associées au chien:")
for successor in G.successors('chien'):
    relation = G.edges[('chien', successor)]['relation']
    print(f"Le chien {relation} {successor}")

# Relation entre "chien" et "animal"
if G.has_edge('chien', 'animal'):
    relation = G.edges[('chien', 'animal')]['relation']
    print(f"\nRelation directe entre chien et animal:")
    print(f"Le chien {relation} animal")
```

## Le résultat montre :
1. Une visualisation graphique du réseau avec tous les concepts et leurs relations
2. La liste des relations associées au concept "chien"
3. La relation spécifique entre "chien" et "animal"
